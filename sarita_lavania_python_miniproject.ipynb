#### Library Management System using Python

```python
# -----------------------------
# Library Management System
# -----------------------------

# Predefined user credentials for login
credentials = {
    "admin": "admin123",
    "librarian": "lib123"
}

# Initial list of books in the system
books = [
    {
        "id": 1,
        "title": "To Kill a Mockingbird",
        "author": "Harper Lee",
        "available": True,
        "issued_to": None
    },
    {
        "id": 2,
        "title": "1984",
        "author": "George Orwell",
        "available": False,
        "issued_to": "Alice"
    },
    {
        "id": 3,
        "title": "The Great Gatsby",
        "author": "F. Scott Fitzgerald",
        "available": True,
        "issued_to": None
    }
]

# -------------------
# Welcome Message
# -------------------
def welcome_message():
    print("=" * 60)
    print("        Welcome to the Library Management System")
    print("=" * 60)
    print("Manage your books easily: view, add, issue, return, and more.\n")

# ---------------------------------------
# Function: Login system (3 attempts max)
# ---------------------------------------
def login():
    attempts = 3
    while attempts > 0:
        username = input("Username: ").strip().lower()  # case-insensitive username
        password = input("Password: ").strip()

        if username in credentials and credentials[username] == password:
            print(f"\nWelcome, {username}! You are logged in.\n")
            return True
        else:
            attempts -= 1
            print(f"Invalid credentials. Attempts left: {attempts}\n")

    print("Too many failed login attempts. Program exiting.")
    return False

# ------------------------------------------
# Function: Display all books with their status
# ------------------------------------------
def display_books():
    print("\nLibrary Books:\n")
    if not books:
        print("No books found in the library.\n")
        return

    for book in books:
        status = "Available" if book["available"] else f"Issued to {book['issued_to']}"
        print(f'ID: {book["id"]} | Title: "{book["title"]}" | Author: {book["author"]} | Status: {status}')
    print()

# ------------------------------------------
# Function: Add a new book if it doesn't exist
# ------------------------------------------
def add_book():
    print("\nAdd a New Book\n")
    title = input("Enter book title: ").strip()
    author = input("Enter author name: ").strip()

    # Check for existing book
    for book in books:
        if book["title"].lower() == title.lower() and book["author"].lower() == author.lower():
            print("\nThis book already exists in the system.\n")
            return

    new_id = max(book["id"] for book in books) + 1 if books else 1
    books.append({
        "id": new_id,
        "title": title,
        "author": author,
        "available": True,
        "issued_to": None
    })

    print(f'\nBook "{title}" by {author} added successfully with ID {new_id}.\n')

# ------------------------------------------
# Function: Issue a book if available
# ------------------------------------------
def issue_book():
    print("\nIssue a Book\n")
    try:
        book_id = int(input("Enter book ID to issue: "))
    except ValueError:
        print("Invalid input! Please enter a number.\n")
        return

    for book in books:
        if book["id"] == book_id:
            if book["available"]:
                person = input("Enter name of person issuing the book: ").strip()
                book["available"] = False
                book["issued_to"] = person
                print(f'\nBook "{book["title"]}" issued to {person}.\n')
                return
            else:
                print(f'\nBook "{book["title"]}" is already issued to {book["issued_to"]}.\n')
                return

    print("Book not found.\n")

# ------------------------------------------
# Function: Return a book if it is issued
# ------------------------------------------
def return_book():
    print("\nReturn a Book\n")
    try:
        book_id = int(input("Enter book ID to return: "))
    except ValueError:
        print("Invalid input! Please enter a number.\n")
        return

    for book in books:
        if book["id"] == book_id:
            if not book["available"]:
                print(f'Book "{book["title"]}" returned from {book["issued_to"]}.\n')
                book["available"] = True
                book["issued_to"] = None
                return
            else:
                print(f'Book "{book["title"]}" was not issued.\n')
                return

    print("Book not found.\n")

# ------------------------------------------
# Function: View all issued books with borrowers
# ------------------------------------------
def view_issued_books():
    print("\nIssued Books:\n")
    issued = [book for book in books if not book["available"]]

    if not issued:
        print("No books are currently issued.\n")
        return

    for book in issued:
        print(f'ID: {book["id"]} | Title: "{book["title"]}" | Issued to: {book["issued_to"]}')
    print()

# ------------------------------------------
# Function: Main menu for user interaction
# ------------------------------------------
def main_menu():
    while True:
        print("Library Management System - Main Menu")
        print("------------------------------------")
        print("1. View Books")
        print("2. Add a Book")
        print("3. Issue a Book")
        print("4. Return a Book")
        print("5. View Issued Books")
        print("6. Exit")
        choice = input("Choose an option (1-6): ").strip()

        if choice == '1':
            display_books()
        elif choice == '2':
            add_book()
        elif choice == '3':
            issue_book()
        elif choice == '4':
            return_book()
        elif choice == '5':
            view_issued_books()
        elif choice == '6':
            print("\nThank you for using the Library Management System. Goodbye!\n")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.\n")

# -----------------------------
# Program Entry Point
# -----------------------------
if __name__ == "__main__":
    welcome_message()
    if login():
        main_menu()




```
